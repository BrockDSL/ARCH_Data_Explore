# Junkin Data Prep

Grabs the last Internet Archive capture of the Mayor's messages page:
https://web.archive.org/web/20210422114625/https://www.pelham.ca/en/living-here/covid-19-messages-from-the-mayor.aspx

Parses it out into CSV file.

```python

#page.html
from bs4 import BeautifulSoup
import csv
import pandas as pd
import requests

URL = "https://web.archive.org/web/20210422114625/https://www.pelham.ca/en/living-here/covid-19-messages-from-the-mayor.aspx"
```

```python
#Grab page
r = requests.get(URL)
with open("page.html", "w") as f:
    f.write(r.text)
```

```python
#Using Soup
soup = BeautifulSoup(open("page.html"),'html.parser')

headers = soup.findAll('td',{"data-name":"accParent"})
entries = soup.findAll('td',{"data-name":"accChild"})

with open('junkin_post.csv','w',newline='') as csvfile:
    my_writer = csv.writer(csvfile)
    for h in headers:
        for e in entries:
            print(h.text)
            print(e.text)
            my_writer.writerow([h.text,e.text])
```

```python
#Parse our and write to disk

header_list = []
headers = soup.findAll('td',{"data-name":"accParent"})
for h in headers:
    header_list.append(h.text.replace("\n",""))
    
entry_list = []
entries = soup.findAll('td',{"data-name":"accChild"})
for e in entries:
    entry_list.append(e.text.replace("\n",""))
    
final_list = pd.DataFrame(list(zip(header_list,entry_list)))
final_list.columns = ["Date","Entry"]
final_list.to_csv("Junkin_Posts.csv",index=False)
```
